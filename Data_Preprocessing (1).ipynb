# Outliers

```python
# Converting data dictionary to DataFrame
df = pd.DataFrame(data)

# Function to detect outliers using IQR method
def detect_outliers(df, columns):
    outliers = pd.DataFrame()
    for col in columns:
        if df[col].dtype in ['int64', 'float64']:  # Check if column is numeric
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            outliers = pd.concat([outliers, col_outliers], axis=0)
    return outliers

columns_to_check = ['Message-ID','Date','From','To','Subject','Message','Cc','Bcc']

outliers_df = detect_outliers(df, columns_to_check)

if not outliers_df.empty:
    print("Outliers detected:")
    print(outliers_df)
else:
    print("No outliers detected.")
```

This code checks for outliers in the given columns and returns the result.

# Min-Max Scaling

```python
df = pd.DataFrame(data)

def min_max_scaling(df):
    for column in df.columns:
        if df[column].dtype in ['int64', 'float64']:  # Check if column is numeric
            min_val = df[column].min()
            max_val = df[column].max()
            df[column] = (df[column] - min_val) / (max_val - min_val)
    return df

df_scaled = min_max_scaling(df)
df_scaled
```

This code performs normalization by scaling the DataFrame to the columns which are numeric.

# Feature Engineering

```python
import pandas as pd
import numpy as np
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
import datetime

enron_df = pd.read_csv('enron.csv')
```

```python
enron_df['email_length'] = enron_df['Message'].apply(len)
enron_df['word_count'] = enron_df['Message'].apply(lambda x: len(x.split()))
enron_df['sentence_count'] = enron_df['Message'].apply(lambda x: len(x.split('.')))
enron_df['avg_word_length'] = enron_df['Message'].apply(lambda x: np.mean([len(word) for word in x.split()]))
```

The above code extracts basic features and adds four new columns to the enron_df DataFrame, calculating the length of each email message, the word count, the sentence count, and the average word length for each message.

```python
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_features = tfidf_vectorizer.fit_transform(enron_df['Message']).toarray()
tfidf_df = pd.DataFrame(tfidf_features, columns=tfidf_vectorizer.get_feature_names_out())
enron_df = pd.concat([enron_df, tfidf_df], axis=1)
```

The code snippet transforms text data into numerical vectors using TF-IDF (Term Frequency-Inverse Document Frequency) by quantifying word importance within each document.TF-IDF (Term Frequency-Inverse Document Frequency) is a method in natural language processing that transforms text into numerical vectors based on the importance of words within each document. It calculates Term Frequency (TF) to measure how often each word appears in a document relative to its total words, giving higher weight to frequently occurring words. Inverse Document Frequency (IDF) adjusts this weight by considering how unique or rare each word is across all documents in the collection of texts, prioritizing rare words that distinguish documents. Together, TF-IDF creates vectors where each dimension represents a word and its weighted importance in the document.

```python
enron_df['sentiment'] = enron_df['Message'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

This code performs sentimental analysis. Sentiment analysis is a field of natural language processing focused on understanding and categorizing the sentiment expressed in textual data. It involves determining whether the sentiment conveyed by a piece of text is positive, negative, or neutral. This analysis is crucial for businesses and organizations to gauge public opinion, assess customer feedback, and make data-driven decisions.
